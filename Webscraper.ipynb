```python
import bs4, requests
res = requests.get('https://www.wuxiaworld.com/novel/stellar-transformations') 
# method to catch a response object from the website 
print(res.status_code)
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
chapters = soup.select('.chapter-item')
chapter_addr = [ ('https://www.wuxiaworld.com'+i.select('a')[0].get('href')) for i in chapters]
chapter_name = [i.select('span')[0].getText() for i in chapters]
```

```python
with open('myNovel.txt','w',encoding = 'UTF-8') as f: 
    #create a new text file called myNovel
    for i in range(len(chapter_addr)): 
        # use for loop to the chapter url and name
        request = requests.get(chapter_addr[i])
        if request.status_code == 200: 
            # judge the connected state, if it is 200, then anlysize again
            soup = soup = bs4.BeautifulSoup(request.text,'lxml')
            content = soup.select('.fr-view')
            a = '' 
            #deal with the text and enable the download keep working
            for j in content:
                b = j.select('p')
                for k in b:
                    a += k.getText()+ '\n'
            f.write(chapter_name[i]) 
            # write Chapter name
            f.write(a)
            # write content
            print('download：',chapter_name[i])
```
